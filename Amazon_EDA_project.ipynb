```python
# Check Missing Values
print(df.isnull().sum())
```

```python
# Remove Duplicate Rows
print(df.duplicated().sum())
df = df.drop_duplicates()
```

```python
# Exploratory Data Analysis (EDA)
# A. Distribution of Ratings
plt.figure(figsize=(8,5))
plt.hist(df["rating"].dropna(), bins=10)
plt.title("Product Ratings Distribution")
plt.xlabel("Rating")
plt.ylabel("Frequency")
plt.show()
```

```python
# Top 10 Categories
top_cat = df["category"].value_counts().head(10)

plt.figure(figsize=(10,5))
plt.bar(top_cat.index, top_cat.values)
plt.xticks(rotation=90)
plt.title("Top 10 Product Categories")
plt.show()
```

```python
# Discounted Price Distribution
plt.figure(figsize=(8,5))
plt.hist(df["discounted_price"].dropna(), bins=20)
plt.title("Discounted Price Distribution")
plt.xlabel("Price")
plt.ylabel("Count")
plt.show()
```

```python
# Actual Price vs Discounted Price
plt.scatter(df["actual_price"], df["discounted_price"])
plt.xlabel("Actual Price")
plt.ylabel("Discounted Price")
plt.title("Actual vs Discounted Price")
plt.show()
```

```python
# Average Rating by Category
avg_rating = (
    df.groupby("category")["rating"]
    .mean()
    .sort_values(ascending=False)
    .head(10)
)

plt.figure(figsize=(10,5))
plt.bar(avg_rating.index, avg_rating.values)
plt.xticks(rotation=90)
plt.title("Top Categories by Average Rating")
plt.show()
```

```python
# Highest Discounted Products
top_discount = (
    df.sort_values("discount_percentage", ascending=False)
      [["product_name", "discount_percentage"]]
      .head(10)
)

print(top_discount)
```

```python
# Correlation Analysis
numeric = df[
    ["actual_price",
     "discounted_price",
     "discount_percentage",
     "rating",
     "rating_count"]
]

print(numeric.corr())
```

```python
# Average Price
avg_price = np.mean(df["discounted_price"].dropna())
print(avg_price)
```

```python
# Median Price
median_price = np.median(df["discounted_price"].dropna())
print(median_price)
```

```python
# Maximum Discount
print(np.max(df["discount_percentage"].dropna()))
```

```python
# Standard Deviation
print(np.std(df["discounted_price"].dropna()))
```

```python
# Top 10 Most Reviewed Products
top_reviewed = (
    df.sort_values("rating_count", ascending=False)
      [["product_name", "rating_count"]]
      .head(10)
)

print(top_reviewed)
```

```python
# Average Discount by Category
avg_discount = (
    df.groupby("category")["discount_percentage"]
      .mean()
      .sort_values(ascending=False)
)

print(avg_discount.head(10))
```

```python
# Top Products by Money Saved
top_saved = (
    df.sort_values("price_saved", ascending=False)
      [["product_name", "price_saved"]]
      .head(10)
)

print(top_saved)
```

```python
# Average Price by Category
avg_price = (
    df.groupby("category")["discounted_price"]
      .mean()
      .sort_values(ascending=False)
)

print(avg_price.head(10))
```

```python
# Rating vs Review Count Scatter Plot
plt.figure(figsize=(8, 5))
plt.scatter(df["rating_count"], df["rating"])
plt.xlabel("Rating Count")
plt.ylabel("Rating")
plt.title("Rating vs Rating Count")
plt.show()
```
